```python
from transformers import pipeline
import gradio as gr
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
def analyze_text(text):
    if text.strip() == "":
        return "Please enter some text!"

    result = classifier(text)[0]
    label = result['label']
    score = round(result['score'], 4)

    return f"Label: {label}\nConfidence Score: {score}"
```

```python
demo = gr.Interface(
    fn=analyze_text,
    inputs=gr.Textbox(lines=3, placeholder="Type your sentence here..."),
    outputs="text",
    title="My First AI Sentiment App",
    description="Type a sentence below to see if it's Positive or Negative!"
)
```

```python
demo.launch(share=True)
```
